```python
import pandas as pd
```

```python
child_csv = pd.read_csv("submission_thresold_0.95.csv")
parent_csv = pd.read_csv("submission_trained_bge_reranker.csv")

mask = (
    child_csv["predicted_citations"].isna()
    | child_csv["predicted_citations"].astype(str).str.strip().str.lower().eq("nan")
)

child_csv.loc[mask, "predicted_citations"] = parent_csv.loc[mask, "predicted_citations"]

output_path = "submission_child_with_parent_backfill.csv"
child_csv.to_csv(output_path, index=False)
print(f"Saved merged submission to {output_path}. Backfilled rows: {mask.sum()}")
```
