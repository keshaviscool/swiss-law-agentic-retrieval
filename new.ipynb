```python
import pandas as pd

val = pd.read_csv("data/val.csv")
```

```python
val.head()
```

```python
cits=[]
for i in val["gold_citations"]:
    for j in i.split(";"):
        # if j not in cits:
        cits.append(j)
```

```python
unique_cits = list(set(cits))
```

```python
unique_cits.__len__()
```

```python
for cit in unique_cits:
    obj = {
        "article": cit,
        "count": cits.count(cit)
    }
    print(obj)
```
