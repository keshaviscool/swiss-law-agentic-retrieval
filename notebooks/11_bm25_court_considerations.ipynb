```python
from rank_bm25 import BM25Okapi
import pandas as pd
import pickle
from tqdm import tqdm
import os
CACHE_DIR = "../indices"
```

```python
# build index
court = pd.read_csv("../data/court_considerations.csv")
```

```python
val = pd.read_csv("../data/val.csv")
```

```python
val["query"][0]
```

```python
texts     = court['text'].fillna('').astype(str).tolist()
citations = court['citation'].astype(str).tolist()
```

```python
tokenized = []
for i, text in enumerate(tqdm(texts, desc="Tokenizing")):
    tokenized.append(text.lower().split())
```

```python
print("Building BM25 index...")
bm25_court = BM25Okapi(tokenized)
print("Done.")
```

```python

os.makedirs(CACHE_DIR, exist_ok=True)
with open(os.path.join(CACHE_DIR, "bm25_court.pkl"), "wb") as f:
    pickle.dump({
        'bm25':      bm25_court,
        'citations': citations,
        'texts':     texts,       # optional but useful for reranking
    }, f)

print(f"Saved. Size: {os.path.getsize(CACHE_DIR)/1e6:.0f} MB")
```

```python
print("Loading BM25 index from cache...")
with open(os.path.join(CACHE_DIR, "bm25_court.pkl"), 'rb') as f:
    cache = pickle.load(f)
```
