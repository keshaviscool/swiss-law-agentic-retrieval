```python
val_csv = pd.read_csv("../data/val.csv")
def _print(index):
    print("query: ", val_csv["query"][index])
    gc = val_csv["gold_citations"][index].split(";")

    result = []

    keys = []

    for i in gc:
        obj = {
            "base": " ".join(i.split(" ")[0:2]),
            "type": "law" if i.split(" ")[0] == "Art." else "court",
            "full": i,
        }
        if obj["base"] not in keys:
            keys.append(obj["base"])
        result.append(obj)

    # print citation by base
    for key in keys:
        print(f"{key}: {[x['full'] for x in result if x['base'] == key]}")
```

```python
train = pd.read_csv("../data/train.csv")
val = pd.read_csv("../data/val.csv")
```

```python
combined = pd.read_csv("../data/combined.csv")
def return_text(citation):
    try:
        return combined[combined["citation"] == citation]["text"].values[0]
    except Exception as e:
        return False
```

```python
print("using untrained model")
untrainted = []
train_idx = 1
gold_citations = val["gold_citations"][train_idx].split(";")
query = val["query"][train_idx+1]
for gc in gold_citations:
    text = return_text(gc)
    if text:
        score = model.predict((query, text))
        print(gc, score)
        untrainted.append([gc, score])

```

```python
val = pd.read_csv("../data/val.csv")
```

```python

print("using trained model")
trained = []
train_idx = 1
gold_citations = val["gold_citations"][train_idx].split(";")
query = val["query"][train_idx+1]
for gc in gold_citations:
    text = return_text(gc)
    if text:
        score = trained_reranker.predict((query, text))
        print(gc, score)
        trained.append([gc, score])

```

### checking delta between trained and untrained

```python
# for correct
for i in range(len(trained)):
    delta = trained[i][1] - untrainted[i][1]
    print(f"Article: {trained[i][0]}, Untrained Score: {untrainted[i][1]}, Trained Score: {trained[i][1]}, Delta: {delta}, percentage change: {(delta/untrainted[i][1])*100:.2f}%")
```

```python
# for wrong
for i in range(len(trained)):
    delta = trained[i][1] - untrainted[i][1]
    print(f"Article: {trained[i][0]}, Untrained Score: {untrainted[i][1]}, Trained Score: {trained[i][1]}, Delta: {delta}, percentage change: {(delta/untrainted[i][1])*100:.2f}%")
```
